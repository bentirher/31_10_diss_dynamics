# Generalization to different emitters

## Introduction

The `QuantumCircuit`s derived in the previous notebook followed from the assumption that both emitters were identical and coupled to the cavity with same strength. This is a very restrictive assumption and so the goal of this notebook is to generalize the two-emitter circuit for non-identical emitters.

## Generalized transformation equations

The normalized transformation equations between the computational and the coupled basis were:

$$ 
\begin{align*}
\ket{G} &= \ket{00} \\
\ket{\Lambda_-} &= - \sin \theta \ket{01} + \cos \theta \ket{10} \\
\ket{\Lambda_+} &= \cos \theta \ket{01} + \sin \theta \ket{10}  \\
\ket{E} &= \ket{11}
\end{align*} 
$$

where $\delta \equiv \tilde{\omega}_0 - \tilde{\omega}_1$, $2\Lambda = \sqrt{4\tilde{g}^2 + \delta^2}$  following the notation in [1](#references). The angle $\theta$ satisfies $\sin \theta = \frac{g}{\sqrt{\Lambda(2 \Lambda + \delta)}} $ and $\cos \theta = \sqrt{\frac{2\Lambda + \delta}{4\Lambda}}$.

This leads to the unitary basis change matrix 

$$ 
\begin{align}
P = \begin{pmatrix}
1 & 0 & 0 & 0 \\
0 & \sin \theta & \cos \theta & 0 \\ 
0 & \cos \theta & - \sin \theta & 0 \\
0 & 0 & 0 & 1
\end{pmatrix}
\end{align}

Since the matrix is symmetric, using $P^{\dag}$ on the coupled basis states leads to the isomorphic transformation equations for the computational basis states

$$ 
\begin{align*}
\ket{00} &= \ket{00} \\
\ket{01} &= - \sin \theta \ket{\Lambda_-} + \cos \theta \ket{\Lambda_+} \\
\ket{10} &= \cos \theta \ket{\Lambda_-} + \sin \theta \ket{\Lambda_+}  \\
\ket{11} &= \ket{E}
\end{align*} 
$$

## Rewriting the dissipators 

First we rewrite the $\sigma_i$ operators into this basis:

$$ 
\begin{align*}
\sigma_0 \equiv \sigma \otimes \mathbf{1} = \ket{0}\bra{1} \otimes (\ket{0}\bra{0} + \ket{1}\bra{1}) = \ket{00} \bra{10} + \ket{01} \bra{11}  = \\
= \ket{G} \left( \cos \theta \bra{\Lambda_-} + \sin \theta \bra{\Lambda_+} \right) + \left( \cos \theta \ket{\Lambda_+} - \sin \theta \ket{\Lambda_-} \right) \bra{E} \equiv \cos \theta \sigma_{\text{G-}} +  \sin \theta \sigma_{\text{G+}} - \sin \theta \sigma_{\text{-E}} + \cos \theta \sigma_{\text{+E}} 
\end{align*} 
$$

$$ 
\begin{align*}
\sigma_1 \equiv \sigma \otimes \mathbf{1} \equiv - \sin \theta \sigma_{\text{G-}} + \cos \theta \sigma_{\text{G+}} + \cos \theta \sigma_{\text{-E}} + \sin \theta \sigma_{\text{+E}} 
\end{align*} 
$$

So the diagonal elements of the dissipators can be expressed in this basis as


$$ 
\begin{align*}
\mathcal{D}_0 [\rho_S] = \tilde{\gamma}_0 \left( \sigma_0 \rho_S \sigma_0^{\dag} - \frac{1}{2} \left\lbrace \sigma_0^{\dag} \sigma_0, \rho_S \right\rbrace \right) \equiv \\
\equiv \tilde{\gamma_0} \left( \cos^2 \theta \mathcal{D}_{\text{G-}} [\rho_S] + \sin^2 \theta \mathcal{D}_{\text{G+}} [\rho_S] +  \sin^2 \theta \mathcal{D}_{\text{-E}} [\rho_S] + \cos^2 \theta \mathcal{D}_{\text{+E}} [\rho_S] + \text{off-diagonal} \right)
\end{align*} 
$$

$$ 
\begin{align*}
\mathcal{D}_1 [\rho_S] = \tilde{\gamma}_1 \left( \sigma_1 \rho_S \sigma_1^{\dag} - \frac{1}{2} \left\lbrace \sigma_1^{\dag} \sigma_1, \rho_S \right\rbrace \right) \equiv \\
\equiv \tilde{\gamma_1} \left( \sin^2 \theta \mathcal{D}_{\text{G-}} [\rho_S] + \cos^2 \theta \mathcal{D}_{\text{G+}} [\rho_S] +  \cos^2 \theta \mathcal{D}_{\text{-E}} [\rho_S] + \sin^2 \theta \mathcal{D}_{\text{+E}} [\rho_S] + \text{off-diagonal} \right)
\end{align*} 
$$

$$ 
\begin{align*}
\mathcal{D}_{01} [\rho_S] + \mathcal{D}_{10} [\rho_S] = \gamma_{01} \left( \sigma_0 \rho_S \sigma_1^{\dag} - \frac{1}{2} \left\lbrace \sigma_1^{\dag} \sigma_0, \rho_S \right\rbrace  + \sigma_1 \rho_S \sigma_0^{\dag} - \frac{1}{2} \left\lbrace \sigma_0^{\dag} \sigma_1, \rho_S \right\rbrace \right) = \\
= 2\tilde{\gamma}_{01} \left( -\sin \theta \cos \theta \mathcal{D}_{\text{G-}} [\rho_S] + \sin \theta \cos \theta \mathcal{D}_{\text{G+}} [\rho_S] - \sin \theta \cos \theta \mathcal{D}_{\text{-E}} [\rho_S] + \sin \theta \cos \theta \mathcal{D}_{\text{+E}} [\rho_S] + \text{off-diagonal} \right) 
\end{align*}
$$

As a result, the diagonal elements of the total dissipator in the generalized coupled basis can be written as 

$$ 
\begin{align*}
\mathcal{D} [\rho_S] = \mathcal{D}_0 [\rho_S] + \mathcal{D}_1 [\rho_S] + \mathcal{D}_{01} [\rho_S] + \mathcal{D}_{10} [\rho_S] \equiv \\
\equiv \left( \tilde{\gamma}_0^2 \cos^2 \theta + \tilde{\gamma}_1^2 \sin^2 \theta - 2 \gamma_{01} \sin \theta \cos \theta \right) \mathcal{D}_{\text{G-}} [\rho_S] + 
\left( \tilde{\gamma}_0^2 \sin^2 \theta + \tilde{\gamma}_1^2 \cos^2 \theta + 2 \gamma_{01} \sin \theta \cos \theta \right) \mathcal{D}_{\text{G+}} [\rho_S] +  \\
+ \left( \tilde{\gamma}_0^2 \sin^2 \theta + \tilde{\gamma}_1^2 \cos^2 \theta - 2 \gamma_{01} \sin \theta \cos \theta \right) \mathcal{D}_{\text{-E}} [\rho_S]  + 
\left( \tilde{\gamma}_0^2 \cos^2 \theta + \tilde{\gamma}_1^2 \sin^2 \theta + 2 \gamma_{01} \sin \theta \cos \theta \right) \mathcal{D}_{\text{+E}} [\rho_S] + \\
+ \text{off-diagonal}
\end{align*} 
$$

We can now define the generalized decay constants $\gamma_{\text{G-}}$, $\gamma_{\text{G+}}$, $\gamma_{\text{-E}}$, $\gamma_{\text{+E}}$

\begin{align*}

\gamma_{\text{G-}} \equiv \tilde{\gamma}_0^2 \cos^2 \theta + \tilde{\gamma}_1^2 \sin^2 \theta - \tilde{\gamma}_{01} \\

\gamma_{\text{G+}} \equiv  \tilde{\gamma}_0^2 \sin^2 \theta + \tilde{\gamma}_1^2 \cos^2 \theta + \tilde{\gamma}_{01} \\

\gamma_{\text{-E}} \equiv \tilde{\gamma}_0^2 \sin^2 \theta + \tilde{\gamma}_1^2 \cos^2 \theta - \tilde{\gamma}_{01} \\

\gamma_{\text{+E}} \equiv \tilde{\gamma}_0^2 \cos^2 \theta + \tilde{\gamma}_1^2 \sin^2 \theta + \tilde{\gamma}_{01}
\end{align*} 

where $\tilde{\gamma}_{01} \equiv 2 \gamma_{01} \sin \theta \cos \theta = \gamma_{01} \sin 2\theta$, so that the dissipator can be written in a more compact way

$$ 
\begin{align*}
\mathcal{D} [\rho_S] = \mathcal{D}_0 [\rho_S] + \mathcal{D}_1 [\rho_S] + \mathcal{D}_{01} [\rho_S] + \mathcal{D}_{10} [\rho_S] \equiv \\
\equiv \gamma_{\text{G-}} \mathcal{D}_{\text{G-}} [\rho_S] + 
\gamma_{\text{G+}} \mathcal{D}_{\text{G+}} [\rho_S] +  
\gamma_{\text{-E}} \mathcal{D}_{\text{-E}} [\rho_S]  + 
\gamma_{\text{+E}} \mathcal{D}_{\text{+E}} [\rho_S] + 
\text{off-diagonal}
\end{align*} 
$$

## Simulations with just the generalized diagonal dissipators

Previously we saw that population dynamics of identical emitters were not affected by off-diagonal dissipators in the coupled basis. This was due to the fact that, even though the expectation values of each emitter population do depend on some off-diagonal elements of the density matrix, these particular elements are not affected by any of the off-diagonal dissipators. On the other hand, if the Hamiltonian evolution were to modify this off-diagonal elements when acting on the system through $U(t) = \exp (-iHt)$, it is not a problem whatsoever since the Hamiltonian dynamics are fully captured (with some approximation error) through the Trotterization of $U(t)$.


Just to check the veracity of this statement, we can do the same check as in the `02_two_emitters_and_one_cavity.ipynb` notebook and verify that with the three different methods (`original`, `markovian`and `diagonal`) we get the same results, now for very different emitters. This generalization can be found in `generalized_ME_methods_comparison.py`

## Generalized quantum circuit for two non-identical emitters

The two key differences from the identical emitters case are: (1) the basis change gate $P$ now depends on the effective coupling $\tilde{g}$ and frequencies $\tilde{\omega}_i$ and (2) instead of just having two angles (reflecting the symmetric decays $E \to \Lambda_{+} \iff \Lambda_{+} \to G$ and $E \to \Lambda_{-} \iff \Lambda_{-} \to G$) in the controlled-Rys, now we have four different angles, each related to one of the decay constants $\gamma_{\text{G+}}, \gamma_{\text{G-}}, \gamma_{\text{-E}}, \gamma_{\text{+E}}$ (or decay probabilities $p_{G+}$, $p_{+E}$, $p_{G-}$, $p_{-E}$).

\begin{align*}
\left| G \right>_S \otimes \left|0 \right>_E &\to \left| G \right>_S \otimes \left|0 \right>_E \\
\left| \Lambda_+ \right>_S \otimes \left|0 \right>_E &\to \sqrt{1- p_{G+}} \left| \Lambda_+ \right>_S \otimes \left|0 \right>_E +  \sqrt{p_{G+}} \left| G \right>_S \otimes \left|1 \right>_E
\end{align*}

\begin{align*}
\left| \Lambda_+ \right>_S \otimes \left|0 \right>_E &\to \left| \Lambda_+  \right>_S \otimes \left|0 \right>_E \\
\left| E \right>_S \otimes \left|0 \right>_E &\to \sqrt{1- p_{+E}} \left| E \right>_S \otimes \left|0 \right>_E +  \sqrt{p_{+E}} \left| \Lambda_+ \right>_S \otimes \left|1 \right>_E
\end{align*}

\begin{align*}
\left| G \right>_S \otimes \left|0 \right>_E &\to \left| G \right>_S \otimes \left|0 \right>_E \\
\left| \Lambda_- \right>_S \otimes \left|0 \right>_E &\to \sqrt{1- p_{G-}} \left| \Lambda_- \right>_S \otimes \left|0 \right>_E +  \sqrt{p_{G-}} \left| G \right>_S \otimes \left|1 \right>_E
\end{align*}

\begin{align*}
\left| \Lambda_- \right>_S \otimes \left|0 \right>_E &\to \left| \Lambda_-  \right>_S \otimes \left|0 \right>_E \\
\left| E \right>_S \otimes \left|0 \right>_E &\to \sqrt{1- p_{-E}} \left| E \right>_S \otimes \left|0 \right>_E +  \sqrt{p_{-E}} \left| \Lambda_- \right>_S \otimes \left|1 \right>_E
\end{align*}

```python
from qiskit import QuantumCircuit, QuantumRegister
from qiskit.circuit import Parameter, ParameterVector, ControlledGate, Gate
from qiskit.quantum_info import Statevector
from qiskit.circuit.library import RYGate
import numpy as np

import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
plt.rcParams['figure.dpi'] = 100
plt.rcParams['savefig.dpi'] = 600

system = QuantumRegister(2, name = 'q')
environment = QuantumRegister(2, name = 'e')

tilde_alpha = Parameter('$\\tilde{\\alpha} (\\tilde{g})$')
tilde_beta = ParameterVector('$\\tilde{\\beta} (\\tilde{\\omega})$', 2)

omega_eff = ParameterVector('$\\tilde{\\omega}$', 2)
g_eff = Parameter('$\\tilde{g}$')

theta_g_plus = Parameter(r'$\theta_{G+}$')
theta_g_minus = Parameter(r'$\theta_{G}$')
theta_minus_e = Parameter(r'$\theta_{E}$')
theta_plus_e = Parameter(r'$\theta_{+E}$')

diff_plus = Parameter(r'$\theta_{E} - \theta_{G+}$')
diff_minus = Parameter(r'$\theta_{+E} - \theta_{G}$')

# Initialization

init = QuantumCircuit(system, environment)
initial_statevector = Statevector([0,0,0,1]) 
init.initialize(params = initial_statevector, qubits = system, normalize = True)

init.barrier()

# Free evolution

u_0 = QuantumCircuit(system, environment)
u_0.rz(tilde_beta[0], system[0])
u_0.rz(tilde_beta[1], system[1])

u_0.barrier()

# Interaction layer

u_int = QuantumCircuit(system, environment)

u_int.rxx(tilde_alpha, qubit1 = system[0], qubit2 = system[1])
u_int.ryy(tilde_alpha, qubit1 = system[0], qubit2 = system[1])

u_int.barrier()

# Decay

u_decay = QuantumCircuit(system, environment)
u_gate = Gate(name = 'P', params = [omega_eff[0], omega_eff[1], g_eff],num_qubits = 2, label = '$P$')
u_dag_gate = Gate(name = 'Pdag', params = [omega_eff[0], omega_eff[1], g_eff], num_qubits = 2, label = '$P^{\\dag}$')
u_decay.append(u_gate, system)
u_decay.barrier()
ccry_plus = ControlledGate(name = 'ryplus', num_qubits = 3, params = [diff_plus], num_ctrl_qubits= 2, base_gate = RYGate(diff_plus))
ccry_minus = ControlledGate(name = 'ryminus', num_qubits = 3, params = [diff_minus], num_ctrl_qubits= 2, base_gate = RYGate(diff_minus))
u_decay.append(ccry_minus, [system[0], system[1], environment[0]])
u_decay.append(ccry_plus, [system[0], system[1], environment[1]])
u_decay.cry(theta_g_minus, system[0], environment[0])
u_decay.cx(environment[0], system[0])
u_decay.cry(theta_g_plus, system[1], environment[1])
u_decay.cx(environment[1], system[1])
u_decay.barrier()
u_decay.append(u_dag_gate, system)
u_decay.reset(environment)

u = init.compose(u_0).compose(u_int).compose(u_decay)

a = u.draw('mpl')
a.savefig('figures/generalized_two_qubit_regular_circuit.png', bbox_inches='tight')
a
```
